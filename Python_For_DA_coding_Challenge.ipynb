```python
#Displaying Missing Values
df.isnull().sum()
```

```python
# Fill missing values in Death_Date and Cause_of_Death
df["Death_Date"] = df["Death_Date"].fillna("Not Applicable")
df["Cause_of_Death"] = df["Cause_of_Death"].fillna("Alive")
```

```python
# Fill missing values in Comorbidities
df["Comorbidities"] = df["Comorbidities"].fillna("None")
```

```python
#Displaying Duplicate Values
df.duplicated().sum()
```

```python
#Finding Unique Values
df.nunique()
```

```python
#Converting Date Columns
df["Diagnosis_Date"] = pd.to_datetime(df["Diagnosis_Date"])
df["Treatment_Start_Date"] = pd.to_datetime(df["Treatment_Start_Date"])
```
